Define Rules and Import Library

```python
#Fewer than two live neighbors -> dead
#Two or three neighbors, lives
#More than three, dies, 
#Three alive becomes alive
import random
```

Functions to Make Life Easier

```python
def pg(grid):
    for row in grid:
        print(row)

def set_state(coords_list, grid):
    for coords in coords_list:
        r = coords[0]
        c = coords[1]
        grid[r][c] = 1
    pg(grid)
    print('-'*len(grid)*2)
    return grid

def randomize(grid):
    dims = len(grid)
    randomized_grid = [[random.choice([0, 1]) for _ in range(dims)] for j in range(dims)]
    return randomized_grid
```

Core Functionality

```python
def set_state(coords_list, grid):
    for coords in coords_list:
        r = coords[0]
        c = coords[1]
        grid[r][c] = 1
    pg(grid)
    print('-'*len(grid)*2)
    return grid


def get_neighbors(i, j, grid):
    positions = [-1, 0, 1]
    neighbors = []
    for row_pos in positions:
        for col_pos in positions:
            if row_pos == 0 and col_pos == 0:
                neighbors.append(0)
            else:
                neighbors.append(grid[i+row_pos][j+col_pos])
    return neighbors



def create_grid(dims):
    return [[0 for j in range(dims)] for i in range(dims)]

def get_status(neighbors, current_state):
    count = sum([int(x) if x != None else 0 for x in neighbors])
    if count < 2: #Die from loneliness
        #print('Died from Loneliness')
        return 0
    if (count == 2 or count == 3) and current_state == 1:
        #print('Stayed alive')
        return 1
    if count > 3 and current_state == 1:
        #print('Died from overpopulation')
        return 0
    if count == 3 and current_state == 0:
        #print('Was born')
        return 1
    return 0


def create_grid(dims):
    return [[0 for j in range(dims)] for i in range(dims)]


def update_grid(grid, generations):
    print('----------------INITIAL STATE----------------')
    pg(grid)
    for num in range(generations):
        for i in range(1, len(grid)-1):
            for j in range(1, len(grid)-1):
                current_state = grid[i][j]
                neighbors = get_neighbors(i, j, grid)
                status = get_status(neighbors, current_state)
                grid[i][j] = status
        print('---------------Generation {}---------------'.format(num+1))
        pg(grid)
    return grid


```

```python
grid = create_grid(15)
updated = randomize(grid)
pg(updated)

new_grid = update_grid(updated, generations=5)
```
